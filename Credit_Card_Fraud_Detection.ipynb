# Credit Card Fraud Detection Using Logistic Regression

This project aims to develop a logistic regression model to detect fraudulent credit card transactions. 

The workflow includes data loading, preprocessing, model training, and evaluation. Given the highly imbalanced nature of the dataset, careful handling is required to ensure accurate detection of fraudulent transactions. 

 Various evaluation metrics and visualizations, such as accuracy, classification reports, confusion matrix, ROC curve, and precision-recall curve, are utilized to assess the model's performance comprehensively.

### Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### loading the dataset to a Pandas DataFrame

```python

credit_card_data = pd.read_csv("creditcard.csv")
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

### This Dataset is highly unblanced

#### 0 --> Normal Transaction

#### 1 --> fraudulent transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

### Under-Sampling

### Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

### Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

### Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

### Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model Training

#### Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

### Model Evaluation

#### Accuracy Score

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

#### 1. Classiification Report
#### 2. Confusion Matrix

```python
# Import necessary libraries for evaluation
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Classification report on test data
print("Classification Report on Test Data:\n")
print(classification_report(Y_test, X_test_prediction))

# Confusion Matrix
cm = confusion_matrix(Y_test, X_test_prediction)

# Visualize Confusion Matrix using heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

#### ROC Curve


```python
# Import necessary libraries for additional plots
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve

# ROC Curve
y_probs = model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class
fpr, tpr, thresholds = roc_curve(Y_test, y_probs)
roc_auc = roc_auc_score(Y_test, y_probs)

plt.figure(figsize=(10, 7))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()


# Print AUC
print(f'Area Under the ROC Curve (AUC): {roc_auc:.2f}')
```

#### Precision-Recall Curve:

```python
# Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(Y_test, y_probs)
plt.figure(figsize=(10, 7))
plt.plot(recall, precision, label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()

```

### Feature Importance

```python
# Feature Importance (for Logistic Regression coefficients)
feature_importance = pd.Series(model.coef_[0], index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 7))
feature_importance.plot(kind='bar')
plt.title('Feature Importance')
plt.show()
```

### Key Observations:
##### V14: Has the largest negative coefficient, indicating it has a strong negative impact on predicting fraud.
##### V4, V5, V7: These features have significant positive coefficients, meaning they are strongly associated with predicting fraud.
##### Amount and Time: These features appear to have less impact on the model compared to some other features.
